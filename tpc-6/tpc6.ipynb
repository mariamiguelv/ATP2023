```python
#Resolução TPC6
file=open("C:/Users/carol/Documents/Eng.Bio2/ATP/datasets/diabetes_prediction_dataset.csv/diabetes_prediction_dataset.csv","r")

def doencasexo(file):
   Fdoenca = 0
   Mdoenca = 0
   for line in file:
       dados = line.split(",")
       doenca = dados[8]
       sexo = dados[0]
       if doenca == "1\n":
           if sexo == "Female":
               Fdoenca =  Fdoenca + 1
           if sexo == "Male":
               Mdoenca =  Mdoenca + 1 
   
   return Fdoenca,Mdoenca


def tabelasexo(Fdoenca,Mdoenca):
    print("|---------------Tabela de doentes por sexo------------------|")
    print("|           Sexo do doente          Distribuição de doentes |")
    print("|              Feminino               " + str(Fdoenca) + "                  |")
    print("|              Masculino              " + str(Mdoenca) + "                  |")
        


   
#Para saber a idade máxima de um paciente doente a considerar na função doencaidade()- 80 anos
#idademax=0
#for line in file:
#      dados = line.split(",")
#      doenca = dados[8]
#      if doenca == "1\n":
#          idade = float(dados[1])
#          if idade >= idademax:
#              idademax = idade
# print(idademax)

def doencaidade(file):
   escalao_idades = [(0,10),(11,24),(25,29),(30,34),(35,39),(40,44),(45,49),(50,54),(55,59),(60,64),(65,69),(70,74),(75,80)]
   distribuicao_escalao = {}
   for escalao in escalao_idades:
        menor , maior = escalao
        distribuicao_escalao[menor,maior]=0
   for line in file:
       dados = line.split(",")
       doenca = dados[8]
       if doenca == "1\n":
           idade = float(dados[1])
           for escalao in escalao_idades:
              menor , maior = escalao
              if menor <= idade <= maior:
                  distribuicao_escalao[menor,maior] = distribuicao_escalao[menor,maior] + 1
   return distribuicao_escalao

def tabelaidade(distribuicao_escalao,escalao_idades):
    print("|---------------Tabela de doentes por idade------------------|")
    print("|         Faixa Etária        Distribuição de doentes        |")
    distribuicao = [n for escalao, n in distribuicao_escalao.items()]
    i = 0
    for escalao in escalao_idades:
       print("|           " + str(escalao) + "                   " + str(distribuicao[i]) + "                 |")
       i = i + 1
         



#Para saber o nível de glucose máximo de um paciente doente a considerar na função glucoselevels()- 300
#glucosemax=0
#for line in file:
      #dados = line.split(",")
      #doenca = dados[8]
      #if doenca == "1\n":
          #glucose = int(dados[7])
          #if glucose >= glucosemax:
              #glucosemax = glucose
#print(glucosemax)

#Para saber o nível de glucose minimo de um paciente doente a considerar na função glucoselevels()- 126
#glucosemin=100000
#for line in file:
      #dados = line.split(",")
      #doenca = dados[8]
      #if doenca == "1\n":
          #glucose = int(dados[7])
          #if glucose <= glucosemin:
              #glucosemin = glucose
#print(glucosemin)

def glucoselevels(file):
    #Criar os intervalos
    min = 126
    max = 0
    intervalos = []
    while max < 300:
        max = min + 9
        if max > 300:
            max = 300 
        intervalo = (min,max)
        intervalos.append(intervalo)
        min= max + 1
    #Criar o dicionário para a contagem de doentes
    distribuicao_intervalos = {}
    for intervalo in intervalos:
       menor , maior = intervalo
       distribuicao_intervalos[menor,maior] = 0
    for line in file:
       dados = line.split(",")
       doenca = dados[8]
       if doenca == "1\n":
           glucoselevel = float(dados[7])
           for intervalo in intervalos:
              menor , maior = intervalo
              if menor <= glucoselevel <= maior:
                  distribuicao_intervalos[menor,maior] = distribuicao_intervalos[menor,maior] + 1
    return distribuicao_intervalos
                  


def tabelaglucose(distribuicao_intervalos,intervalos):
    print("|---------------Tabela com os níveis de glucose dos doentes------------------|")
    print("|            Níveis de glucose            Distribuição de doentes            |")
    distribuicao = [n for intervalo, n in distribuicao_intervalos.items()]
    i = 0
    for intervalo in intervalos:
       print("|              " + str(intervalo) + "                          " + str(distribuicao[i]) + "                      |")
       i = i + 1
        


def menu():
    print(""" 
    (1) Distribuição de doentes com diabetes por sexo
    (2) Distribuição de doentes com diabetes por escalões etários
    (3) Distribuição de doentes com diabetes por níveis de glucose 
    (0) Sair
    """)




```
